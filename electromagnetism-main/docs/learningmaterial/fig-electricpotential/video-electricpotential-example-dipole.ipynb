# Elektrisk dipol

$$V(\vec{r})= \sum_i \frac{Q_i}{4 \pi \epsilon_0 R_i}$$

```python
import numpy as np
import matplotlib.pyplot as plt

def epotlist(r,Q,R):
    V = 0
    for i in range(len(R)):
        Ri = r - R[i]
        Qi = Q[i]
        Rinorm = np.linalg.norm(Ri)
        V = V + Qi/Rinorm
    return V
```

```python
# Sette opp ladninger i liste
R = []
Q = []
r1 = np.array([1,0])
Q1 = 1
R.append(r1)
Q.append(Q1)
r2 = np.array([-1,0])
Q2 = -1
R.append(r2)
Q.append(Q2)
```

```python
Lx = 3
Ly = 3
N = 21
x = np.linspace(-Lx,Lx,N)
y = np.linspace(-Ly,Ly,N)
rx,ry = np.meshgrid(x,y)
```

```python
# Finne verdier for elektrisk potensiale p√• grid
V = np.zeros((N,N),float)
for i in range(len(rx.flat)):
    r = np.array([rx.flat[i],ry.flat[i]])
    V.flat[i] = epotlist(r,Q,R)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(V)
plt.colorbar()
```

```python
plt.figure(figsize=(8,8))
plt.contourf(rx,ry,V,20)
plt.colorbar()
```

# Elektrisk felt

$$\vec{E} = - \nabla V$$

```python
Ey,Ex = np.gradient(V)
Ey = -Ey
Ex = -Ex
```

```python
plt.figure(figsize=(8,8))
plt.contourf(rx,ry,V,20)
plt.quiver(rx,ry,Ex,Ey)
```
