# Ringladning

```python
import numpy as np
import matplotlib.pyplot as plt

def epotlist(r,Q,R):
    V = 0
    for i in range(len(R)):
        Ri = r - R[i]
        Qi = Q[i]
        Rinorm = np.linalg.norm(Ri)
        V = V + Qi/Rinorm
    return V
```

```python
# Lage ring
Q = []
R = []
M = 100
a = 2.0
q = 1.0
deltatheta = 2*np.pi/M
for i in range(M):
    thetai = i*deltatheta
    xi = a*np.cos(thetai)
    yi = a*np.sin(thetai)
    ri = np.array([xi,yi,0])
    R.append(ri)
    Qi = q/M
    Q.append(Qi)
```

```python
Lx = 5
Lz = 5
N = 30
x = np.linspace(-Lx,Lx,N)
z = np.linspace(-Lz,Lz,N)
rx,rz = np.meshgrid(x,z)
```

```python
V = np.zeros((N,N),float)
for i in range(len(rx.flat)):
    r = np.array([rx.flat[i],0,rz.flat[i]])
    V.flat[i] = epotlist(r,Q,R)
```

```python
plt.figure(figsize=(8,8))
plt.contourf(V)
```

```python
Ez,Ex = np.gradient(-V)
plt.figure(figsize=(8,8))
plt.contourf(rx,rz,V)
plt.quiver(rx,rz,Ex,Ez)
```
