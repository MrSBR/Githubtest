```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def efield(r,qi,ri):
    # 1/(4*pi*epsilon_0) = 1
    Ri = r - ri
    Rinorm = np.linalg.norm(Ri)
    return qi*Ri/Rinorm**3
```

```python
q1 = 1.0
r1 = np.array([0,0])
r = np.array([2,0])
print(efield(r,q1,r1))
```

```python
L = 2
N = 10
x = np.linspace(-L,L,N)
y = np.linspace(-L,L,N)
rx,ry = np.meshgrid(x,y)
```

```python
print(rx)
```

```python
print(ry)
```

```python
plt.quiver(rx,ry,rx,ry)
plt.axis('equal')
```

```python
Ex = np.zeros((N,N),float)
Ey = np.zeros((N,N),float)
for i in range(len(rx.flat)):
    r = np.array([rx.flat[i],ry.flat[i]])
    Ex.flat[i],Ey.flat[i] = efield(r,q1,r1)
```

```python
plt.quiver(rx,ry,Ex,Ey)
```

```python
L = 2
N = 30
x = np.linspace(-L,L,N)
y = np.linspace(-L,L,N)
rx,ry = np.meshgrid(x,y)
q1 = -1.0
r1 = np.array([0,0])
q2 = 1.0
r2 = np.array([1,0])
Ex = np.zeros((N,N),float)
Ey = np.zeros((N,N),float)
for i in range(len(rx.flat)):
    r = np.array([rx.flat[i],ry.flat[i]])
    Ex.flat[i],Ey.flat[i] = efield(r,q1,r1) + efield(r,q2,r2)
plt.figure(figsize=(10,10))
plt.quiver(rx,ry,Ex,Ey)
plt.axis('equal')
```
